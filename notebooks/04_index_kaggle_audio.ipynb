```python
from pathlib import Path
import pandas as pd

# Figure out project root (same trick as before)
cwd = Path.cwd()
BASE_DIR = cwd.parent if cwd.name.lower() == "notebooks" else cwd

root = BASE_DIR / "data" / "raw_audio" / "PhishingVoiceDataset"

print("Project root:", BASE_DIR)
print("Audio root:", root)
print("Exists?", root.exists())

rows = []

# Map folder names to our labels
label_map = {
    "Phishing": "vishing",
    "NonPhishing": "safe",
}

for folder_name, label in label_map.items():
    folder = root / folder_name
    print(f"Scanning {folder} ...")
    for audio_path in sorted(folder.glob("*.mp3")):
        rows.append(
            {
                "filename": audio_path.name,
                "label": label,
                "source": "kaggle_phishingvoice",
                # relative path from project root (handy later)
                "rel_path": str(audio_path.relative_to(BASE_DIR)),
            }
        )

df = pd.DataFrame(rows)
out_path = BASE_DIR / "data" / "kaggle_voice_index.csv"
df.to_csv(out_path, index=False, encoding="utf-8-sig")

print("\n✅ Index CSV saved to:", out_path.resolve())
print("Total audio files indexed:", len(df))
df.head()
```
