```python
import pandas as pd
import joblib
from collections import Counter

# Adjust the path since you're in the notebooks folder
df = pd.read_csv("../data/cleaned_dataset.csv")
print("Label counts:", Counter(df["label"]))

model = joblib.load("../models/svm_model.pkl")

# Use the correct column name
df["text"] = df["transcript"].astype(str)

print("✅ Data and model loaded successfully.")
```

```python
safe_samples = df[df["label"] == "safe"].sample(10, random_state=42)

print("=== SAFE SAMPLES ===\n")
for t in safe_samples["text"]:
    print("TEXT:", t[:120])
    print("PRED:", model.predict([t])[0])
    print("-"*60)
```

```python
vishing_samples = df[df["label"] == "vishing"].sample(10, random_state=42)

print("=== VISHING SAMPLES ===\n")
for t in vishing_samples["text"]:
    print("TEXT:", t[:120])
    print("PRED:", model.predict([t])[0])
    print("-"*60)
```

```python
from sklearn.metrics import classification_report

X = df["text"].tolist()
y_true = df["label"].tolist()
y_pred = model.predict(X)

print(classification_report(y_true, y_pred))
```
